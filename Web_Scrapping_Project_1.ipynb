## Importing Necessary Library

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

## Scrapping Data from Web Page

 ### Step 1
         

Passing web page link from which we scrap data.[link text](https://www.senate.gov/legislative/LIS/roll_call_votes/vote1171/vote_117_1_00003.htm#top)

```python
data = requests.get("https://www.senate.gov/legislative/LIS/roll_call_votes/vote1171/vote_117_1_00003.htm#top")
soup = bs(data.content)     # Extracting HTML code of whole web page
p = soup.prettify           # Removing other unwanted things
```

### Step 2

Choosing best  '***class***'  to scrap data

```python
i = soup.find_all(class_="contenttext responsive_col")  # From the HTML code we choose the class = "contenttext responsive_col" to scrap data.
```

### Step 3

Scrapped Data are not well organized and have some unwanted symbols and word.

So, After Scrapping, there might be need of cleaning data and Organized it properly.

```python
# Scrapping data from HTML code
l = []
for j in i:
  l.append(j.text)
print(l)
```

```python
# Oraganizing data

Home_State = []
First = []
Second = []
for i in range(len(l)):
  if i%3==0:
    Home_State.append(l[i])
  elif (i)%3==1:
    First.append(l[i])
  elif (i)%3==2:
    Second.append(l[i])
```

```python
for j in range(len(Home_State)):
  Home_State[j] = Home_State[j][1:-3]
```

```python
for j in range(len(First)):
  First[j] = First[j][:-1]
```

```python
for j in range(len(Home_State)):
  Second[j] = Second[j][:-1]
```

```python
First_member_name , Other = [] , []
for i in First :
  member_name , other = i.split('(')
  First_member_name.append(member_name)
  Other.append(other)
```

```python
First_Member_Party , First_member_vote = [],[]
for i in Other:
  Member_Party , Other_2 = i.split('),')
  First_Member_Party.append(Member_Party)
  First_member_vote.append(Other_2)
```

```python
Second_member_name , Other = [] , []
for i in Second :
  member_name , other = i.split('(')
  Second_member_name.append(member_name)
  Other.append(other)
```

```python
Second_Member_Party , Second_member_vote = [],[]
for i in Other:
  Member_Party , Other_2 = i.split('),')
  Second_Member_Party.append(Member_Party)
  Second_member_vote.append(Other_2)
```

### Step4

After extracting data and organised. Generally, we set our extract data in dataframe

```python
Data = pd.DataFrame({'HOME STATE':Home_State,'First Member Name':First_member_name,'First_Member_Party':First_Member_Party,'First_Member_Vote':First_member_vote,'Second Member Name':Second_member_name,'Second_Member_Party':Second_Member_Party,'Second_Member_Vote':Second_member_vote})
```
